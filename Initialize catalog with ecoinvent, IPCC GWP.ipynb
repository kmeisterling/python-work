## STEP 2: add ecoinvent
the `EcoinventConfig` class just takes a directory name, and expects the ecospold xml files to be saved according to a predictable directory naming scheme beneath that.

```python
ec = EcoinventConfig('/Users/Kyle/lca-data-repo/ecoinvent/')
for r in ec.references:
    for k in ec.make_resources(r):
        cat.add_resource(k)
        
```
